# Tic Tac Toe

#Two players game

```python
# make empty board with 9 space
board = [" "] * 9
# print to show board
def print_board():
    print()
    print(board[0], "|", board[1], "|", board[2])
    print("--+---+--")
    print(board[3], "|", board[4], "|", board[5])
    print("--+---+--")
    print(board[6], "|", board[7], "|", board[8])
    print()

# function to check who is winner 
def check_winner(player):
    # all possible winning lines (rows, coulmns, diagonals)
    wins = [(0,1,2),(3,4,5),(6,7,8),
            (0,3,6),(1,4,7),(2,5,8),
            (0,4,8),(2,4,6)]
            # check each line if same player is in all 3 postions than winner
    for a,b,c in wins:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

# game start here
turn = "X"   # first turn is for "X"
for i in range(9):    # maximum 9 moves because 9 boxes
    print_board()
    # ask playe to give postion number 0 to 8
    pos = int(input(f"Player {turn}, enter position (0-8): "))
    # check postion is empty
    if board[pos] == " ":
        # put players mark "X & 0"
        board[pos] = turn
        # check this move makes the player winner
        if check_winner(turn):
            print_board()
            print("Player", turn, "wins!") # print winner
            break
        # if not win than change the turn "X to 0 & 0 to X"
        turn = "O" if turn == "X" else "X"
        # if place already filled than ask again 
    else:
        print("Position already taken! Try again.")
else:
    # if all 9 moves done and no winner than print game is draw
    print_board()
    print("game is Draw!")
```
